# Normalización | Preprocesamiento de datos | SciKitLearn | Machine Learning

 La normalización es un proceso de preprocesamiento que escala los datos numéricos a un rango específico, generalmente entre 0 y 1 o entre -1 y 1. Esto se hace para asegurar que todas las características tengan un impacto similar en el modelo de aprendizaje automático, especialmente cuando las escalas de las características varían considerablemente

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing
```

```python
datos = pd.read_csv("movies.csv")
datos = datos.replace(np.nan,"0")
df = pd.DataFrame(datos)

print("Media: ")
print(df["movie_facebook_likes"].mean())

print("Desviacion Estandar: ")
print(df["movie_facebook_likes"].std())

print("Valor maximo: ")
print(df["movie_facebook_likes"].max())

print("Valor minimo: ")
print(df["movie_facebook_likes"].min())

fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(10,5))

ax1.set_title("Antes de escalar")
sns.kdeplot(df["movie_facebook_likes"],ax=ax1,color="red")
sns.kdeplot(df["imdb_score"],ax=ax1)

# scaler=preprocessing.StandardScaler()
scaler=preprocessing.Normalizer(norm="l2",copy=True)

df[["imdb_score","movie_facebook_likes"]]=scaler.fit_transform(df[["imdb_score","movie_facebook_likes"]])

print("*"*20)

print(df["movie_facebook_likes"].iloc[0])
print("Media: ")
print(df["movie_facebook_likes"].mean())

print("Desviacion Estandar: ")
print(df["movie_facebook_likes"].std())

print("Valor maximo: ")
print(df["movie_facebook_likes"].max())

print("Valor minimo: ")
print(df["movie_facebook_likes"].min())

df.to_csv("prepro2.csv",sep="\t")

ax2.set_title("Despues de escalar")
sns.kdeplot(df["movie_facebook_likes"],ax=ax2)
sns.kdeplot(df["imdb_score"],ax=ax2)

plt.show()

```
