# Agrupacion en Clustering (K-means)
## ¬øQue es K-means?

K-means es un algoritmo de clasificaci√≥n no supervisada (clusterizaci√≥n) que agrupa objetos en k grupos bas√°ndose en sus caracter√≠sticas. El agrupamiento se realiza minimizando la suma de distancias entre cada objeto y el centroide de su grupo o cluster. Se suele usar la distancia cuadr√°tica.

# üé¨ Descripci√≥n del problema: Agrupaci√≥n de pel√≠culas con K-Means

El prop√≥sito de este an√°lisis es agrupar pel√≠culas similares en funci√≥n de dos caracter√≠sticas importantes:

1. cast_total_facebook_likes: n√∫mero total de likes en Facebook del reparto (actores y actrices).

2. imdb_score: puntuaci√≥n promedio de la pel√≠cula en IMDb (Internet Movie Database), de 1 a 10.

Estas dos variables nos permiten obtener una idea general de:

* Cu√°n popular es el reparto de una pel√≠cula.

* Qu√© tan buena es considerada la pel√≠cula seg√∫n los usuarios.


## ü§ñ ¬øQu√© hace el algoritmo?

Usamos K-Means, un algoritmo de agrupamiento no supervisado, para agrupar pel√≠culas que sean similares entre s√≠ bas√°ndose en esos dos criterios.

* No sabemos de antemano a qu√© grupo pertenece cada pel√≠cula.

* El algoritmo encuentra patrones naturales y divide las pel√≠culas en clusters (grupos).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#Leer archivo desde el escritorio (Google Colab)
from google.colab import files
uploaded = files.upload()
```

```python
datos = pd.read_csv('moviescs.csv')
df = pd.DataFrame(datos)

#Extrar valores
x = df['cast_total_facebook_likes'].values
y = df['imdb_score'].values

#Manejar lo datos
print("Valor maximo de likes: ",df['cast_total_facebook_likes'].max())
print("Valor minimo de likes: ",df['cast_total_facebook_likes'].min())
print("Valor promedio de likes: ",df['cast_total_facebook_likes'].mean())
```

```python
# Crear la matriz de datos
# info = df[['cast_total_facebook_likes', 'imdb_score']].values
# print(info)
X = np.array(list(zip(x,y)))
print(X)
```

```python
kmeans = KMeans(n_clusters=3) # Cambiar el numero de agrupaciones
labels = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_

colors = ["m.","r.","c","y.","b."]

# Imprimir etiquetas asignadas
for i in range (len(x)):
  print("Cordenada: ",X[i]," - Grupo: ",labels[i])
  plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize=10)


# Graficar los centroides
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x',
            s=150, linewidths=2, zorder = 10, c='yellow', label='Centroides')
plt.title("Clustering de Pel√≠culas por Likes y Puntuaci√≥n Imdb")
plt.xlabel("Likes en Facebook (cast_total_facebook_likes)")
plt.ylabel("Puntuaci√≥n IMDb (imdb_score)")
plt.legend()
plt.grid(True)
plt.show()

```
