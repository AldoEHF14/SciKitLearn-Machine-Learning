# Validaci√≥n cruzada | K-Fold
## üß™ ¬øQu√© es la validaci√≥n cruzada?
La validaci√≥n cruzada es una t√©cnica que usamos para evaluar qu√© tan bien se desempe√±a un modelo de machine learning sin usar un conjunto de prueba fijo. Nos ayuda a saber si el modelo generaliza bien a nuevos datos, o si solo "memoriz√≥" los datos de entrenamiento.

## üîÅ ¬øY qu√© es K-Fold Cross Validation?
K-Fold (o "K-particiones") es una forma espec√≠fica de validaci√≥n cruzada.

üìå Paso a paso:
Se divide el conjunto de datos en K partes iguales (llamadas folds).

1. El modelo se entrena K veces, cada vez con un fold diferente como conjunto de prueba y los dem√°s como entrenamiento.

2. Se calculan las m√©tricas (por ejemplo, precisi√≥n, error, R¬≤) en cada uno de los K folds.

3. Al final, se promedia el resultado de todas las evaluaciones. Esto nos da una idea m√°s confiable del rendimiento del modelo.

## üéØ ¬øPor qu√© se usa?
* Para evitar el sobreajuste (overfitting).

* Para aprovechar todos los datos disponibles tanto para entrenamiento como para prueba.

* Para obtener una evaluaci√≥n m√°s estable y justa del modelo.

## üìå Ejemplo simple (K=5)
Supongamos que tienes 100 datos. Si haces K-Fold con K=5:

Divides tus datos en 5 grupos de 20.

En la primera iteraci√≥n, usas el grupo 1 como prueba y los grupos 2‚Äì5 como entrenamiento.

En la segunda iteraci√≥n, usas el grupo 2 como prueba y los otros para entrenamiento.

Y as√≠ hasta que cada grupo haya sido usado una vez como prueba.

Mostraremos como utilizar la herramienta de K-Fold  que se utiliza para hacer la validaci√≥n cruzada, aqu√≠ te mostramos c√≥mo hacerlo a partir de datos de pel√≠culas de kaggle.

```python
from sklearn.model_selection import KFold # Dividir el conjunto
import pandas as pd
from sklearn import linear_model # Modelo de regresi√≥n lineal.
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
# Leer los datos
datos = pd.read_csv('moviesc10.csv')
df = pd.DataFrame(datos)

# Definir variables
X = df["cast_total_facebook_likes"].values.reshape(-1, 1)  # Variable independiente
y = df["imdb_score"].values # Variable dependiente

# Separacion de los datos, pero basandose en el indice (reparticion de los datos)
# Configurar validaci√≥n cruzada con 2 divisiones
kf=KFold(n_splits=2,shuffle =True,random_state=2)

for valores_x,valores_y in kf.split(X):
  # print("Entrenamiento: ",df.iloc[valores_x],"Prueba: ",df.iloc[valores_y])
  print(valores_x,valores_y)

# Modelo de regresion lineal
regr = linear_model.LinearRegression()


for entrenamiento_indice,prueba_indice in kf.split(X):
  print("\nEntrenamiento: ",entrenamiento_indice,"Prueba: ",prueba_indice)
  X_entrenamiento,X_prueba = X[entrenamiento_indice],X[prueba_indice]
  y_entrenamiento,y_prueba = y[entrenamiento_indice],y[prueba_indice]
  regr.fit(X_entrenamiento,y_entrenamiento)
  y_pred = regr.predict(X_prueba)
  print(y_pred)
  print("Error absoluto medio:", mean_absolute_error(y_prueba, y_pred))
  print("Coeficiente R¬≤:", r2_score(y_prueba, y_pred))

```
